## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")

# YOUR CODE HERE
# Read in data
whale_returns =  pd.read_csv(whale_returns_csv, index_col = 'Date', parse_dates=True, infer_datetime_format=True)
whale_returns.head()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
# YOUR CODE HERE
algo_returns = pd.read_csv(algo_returns_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)

algo_returns.head()
```

```python
# Count nulls
algo_returns.isnull().sum()
# YOUR CODE HERE
```

```python
# Drop nulls
algo_returns = algo_returns.dropna()
# YOUR CODE HERE
algo_returns.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices, sorting index
sp500_history_csv = Path("Resources/sp500_history.csv")
# YOUR CODE HERE
sp500_history = pd.read_csv(sp500_history_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
sp500_history.sort_index(inplace=True)
sp500_history.head()
```

```python
# Check Data Types
sp500_history.dtypes
# YOUR CODE HERE
```

```python
#Fix datatype
sp500_history['Close'] = sp500_history['Close'].replace(r'[$,]','',regex=True).astype('float64')
sp500_history.head()
```

```python
# Calculate Daily Returns
# YOUR CODE HERE
sp500_daily_returns = sp500_history.pct_change()
sp500_daily_returns.head()
```

```python
# Drop nulls
# YOUR CODE HERE
sp500_daily_returns.isnull().sum()
sp500_daily_returns.dropna(inplace=True)
```

```python
# Rename Column
# YOUR CODE HERE
sp500_daily_returns.columns = ['S&P 500']
sp500_daily_returns.head()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame

all_portfolio_returns = pd.concat([whale_returns, algo_returns, sp500_daily_returns], axis='columns', join='inner')
all_portfolio_returns.head()
```

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
# YOUR CODE HERE
all_portfolio_returns.plot.box(figsize=[20,10])
```

```python
# Daily Standard Deviations

# Calculate the standard deviation for each portfolio. Which portfolios are riskier than the S&P 500?

portfolio_std = all_portfolio_returns.std()
portfolio_std
```

```python
# Calculate the annualized standard deviation (252 trading days)

annualized_std = portfolio_std * np.sqrt(252)
annualized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
2. Calculate and plot a 60 day Beta for Berkshire Hathaway Inc compared to the S&&P 500

```python
# Calculate and plot the rolling standard deviation for the S&PP 500 using a 21 day window
all_portfolio_returns.rolling(window=21).std().plot(figsize=[20,10])
```

```python
# Correlation
# YOUR CODE HERE
all_portfolio_returns.corr()
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
covar_berkshire = all_portfolio_returns['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(all_portfolio_returns['S&P 500'].rolling(window=60))

# Calculate variance of all daily returns of stocks vs. S&P 500
variance = all_portfolio_returns['S&P 500'].rolling(window=60).var()

# Calculate beta of all daily returns of stocks
berkshire_beta = covar_berkshire / variance
berkshire_beta
```

```python
berkshire_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of Berkshire Hathaway Inc')
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

Calculate and plot the annualized Sharpe ratios for all portfolios to determine which portfolio has the best performance

```python
# Annualzied Sharpe Ratios
sharpe_ratios = (all_portfolio_returns.mean() * 252) / (all_portfolio_returns.std() * np.sqrt(252))
sharpe_ratios
```

 plot() these sharpe ratios using a barplot.
 On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales?

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(figsize=[15,10], kind="bar", title="Sharpe Ratios")
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
aapl_stock_csv = Path("Resources/aapl_historical.csv")

# Read in data
aapl_returns =  pd.read_csv(aapl_stock_csv, index_col = 'Trade DATE', parse_dates=True, infer_datetime_format=True, usecols=['Trade DATE', 'NOCP'])
aapl_returns.sort_index(inplace=True)
aapl_returns.head()
```

```python
# Read the second stock
google_stock_csv = Path("Resources/goog_historical.csv")

# Read in data
goog_returns =  pd.read_csv(google_stock_csv, index_col = 'Trade DATE', parse_dates=True, infer_datetime_format=True, usecols=['Trade DATE', 'NOCP'])
goog_returns.sort_index(inplace=True)
goog_returns.head()
```

```python
# Read the third stock
cost_stock_csv = Path("Resources/cost_historical.csv")

# Read in data
cost_returns =  pd.read_csv(cost_stock_csv, index_col = 'Trade DATE', parse_dates=True, infer_datetime_format=True, usecols=['Trade DATE', 'NOCP'])
cost_returns.sort_index(inplace=True)
cost_returns.head()
```

```python
# Concatenate all stocks into a single DataFrame
my_portfolios_df = pd.concat([aapl_returns, goog_returns, cost_returns], axis='columns', join='inner')

my_portfolios_df.columns=['AAPL', 'GOOG', 'COST']
my_portfolios_df
```

```python
# calculate daily returns
# YOUR CODE HERE
my_portfolio_daily_returns = my_portfolios_df.pct_change()
my_portfolio_daily_returns.head()
```

```python
# Drop Nulls
# YOUR CODE HERE
my_portfolio_daily_returns.dropna(inplace=True)
my_portfolio_daily_returns
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
# YOUR CODE HERE
weighted_portfolio_returns = my_portfolio_daily_returns.dot(weights)
#weighted_portfolio_returns = weighted_portfolio_returns.resample('D').mean().dropna()
weighted_portfolio_returns.sum()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# YOUR CODE HERE
combined_portfolios = pd.concat([whale_returns, sp500_daily_returns, weighted_portfolio_returns], axis='columns', join='inner')
combined_portfolios.rename(columns={0:'MY PORTFOLIO'}, inplace=True)
combined_portfolios.head() 
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
combined_portfolios.dropna(inplace=True)
combined_portfolios
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Risk
combined_portfolios_std = combined_portfolios.std() * np.sqrt(252)
combined_portfolios_std
```

```python
# Rolling
combined_portfolios.rolling(window=30).std()
```

```python
# Beta
covar_my_portfolio = combined_portfolios['MY PORTFOLIO'].cov(combined_portfolios['S&P 500'])

# Calculate variance of all daily returns of stocks vs. S&P 500
variance = combined_portfolios['S&P 500'].var()

# Calculate beta of all daily returns of stocks
my_portfolio_beta = covar_my_portfolio / variance
my_portfolio_beta
```

```python
# Annualzied Sharpe Ratios
sharpe_ratios_my_portfolio = (combined_portfolios.mean() * 252) / (combined_portfolios.std() * np.sqrt(252))
sharpe_ratios_my_portfolio
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios_my_portfolio.plot(figsize=[15,10], kind="bar", title="Sharpe Ratios My Portfolio")
```
